create new column with marks + 1 and also update existing dataframe

```python
df.withColumn('New_Marks',df.Marks+1).show()
```

rename name column and give new name "Student_name"

```python
df.withColumnRenamed('Name','Student_name').show()
```

display name of the student having marks greater than 90

```python
df.filter(df['Marks']>90).select('Name').show()
```

display name of the student having marks greater than 90

```python
df.filter((df['Marks']<90) & (df['Gender']=='Female')).select('Name').show()
```

display name of the male student having marks greater than 90

```python
df.filter((df['Marks']>90) & ~(df['Gender']=='Female')).select('Name').show()
```

display avg marks

```python
df.groupBy('Gender').mean().select(['Gender','avg(Marks)']).show()
```

sort every row of the dataset into descending order

```python
df.orderBy(df['Marks']).show()
```
